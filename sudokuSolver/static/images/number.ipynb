```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
one = cv2.imread("numbers/one.png")
two = cv2.imread("numbers/two.png")
three = cv2.imread("numbers/three.png")
four = cv2.imread("numbers/four.png")
five = cv2.imread("numbers/five.png")
six = cv2.imread("numbers/six.png")
seven = cv2.imread("numbers/seven.png")
eight = cv2.imread("numbers/eight.png")
nine = cv2.imread("numbers/nine.png")

numbers = [one, two, three, four, five, six, seven, eight, nine]
```

```python
number_names = ["one", "two", "three", "four", "five", "six", "seven", "eight", "nine"]

for i, num in enumerate(numbers):
    num_gray = cv2.cvtColor(num, cv2.COLOR_BGR2GRAY)
    num_mask = 255 - np.array(num_gray > 50) * 255

    num_bg = cv2.cvtColor(num, cv2.COLOR_BGR2BGRA)
    num_bg[..., 0] = np.zeros((200, 200))
    num_bg[..., 1] = num_mask
    num_bg[..., 2] = np.zeros((200, 200))
    num_bg[..., 3] = num_mask

    num_bg = cv2.resize(num_bg, (100, 100))
    cv2.imwrite(f"numbers/{number_names[i]}.png", num_bg)
```
